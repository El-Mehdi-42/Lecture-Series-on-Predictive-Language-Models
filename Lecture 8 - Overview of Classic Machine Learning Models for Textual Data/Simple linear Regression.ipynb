# Importing packages

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Taille de l'échantillon (souhaitée)
n = 200
```

```python
X = np.random.randint(1,30,n)
X
```

```python
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X,3) + 3*np.ones(n) + 6*np.random.randn(n)
```

```python
plt.plot(y,X,"o")
plt.show()
```

```python
reg = LinearRegression()
reg.fit(np.c_[X, np.ones(n)],y)
```

```python
pente = reg.coef_[0]
pente
```

```python
cst = reg.intercept_
cst
```

```python
import matplotlib.pyplot as plt
plt.plot(y,X,"o")
plt.plot(pente*X+cst*np.ones(n),X)
plt.show()
```

```python
reg.predict([[1,10],[2,25]])
```

```python
# La deuxième coordonnée est muette, mais il faut mettre quand même qqch sinon il ne va pas tourner
reg.predict(np.array([[3, 1000]]))
```

```python
reg.predict(np.array([[15, 1000]]))
```

```python
import matplotlib.pyplot as plt
plt.plot(y,X,"o")
plt.plot(pente*X+cst*np.ones(n),X)
plt.scatter(reg.predict([[3,5], [15, 111]]),[3, 15], marker="*", color="r", linewidths=4)
plt.show()
```

```python
score = reg.score(np.c_[X,np.ones(n)],y)
score
```

```python
print("Précision: ", round(score*100,3),"%")
```
