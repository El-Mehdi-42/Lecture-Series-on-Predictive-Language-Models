```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000, n_features=2, centers=2, random_state=40)
y = y.reshape((y.shape[0], 1))

print('dimensions de X:', X.shape)
print('dimensions de y:', y.shape)

fig, ax = plt.subplots(figsize=(9, 6))

#plt.style.use('dark_background')
#plt.style.use('Solarize_Light2')

# fig.suptitle("Graphic")
ax.scatter(X[:,0], X[:, 1], c=y, cmap="copper")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_test.shape
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
pred_y = classifier.predict(X_test)
print(pred_y[10:])
print(Y_test)
accuracy = classifier.score(X_test,Y_test)
print(accuracy)
```

```python
classifier = DecisionTreeClassifier(criterion='gini',random_state=0)
classifier.fit(X_train,Y_train)
pred_y = classifier.predict(X_test)
accuracy = classifier.score(X_test,Y_test)
print(accuracy)
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', min_samples_split=30,random_state=0)
classifier.fit(X_train,Y_train)
pred_y = classifier.predict(X_test)
accuracy = classifier.score(X_test,Y_test)
print(accuracy)
```

```python
a = []
for i in range(2,40):
    classifier = DecisionTreeClassifier(criterion='entropy', min_samples_split=i, random_state=0)
    classifier.fit(X_train,Y_train)
    pred_y = classifier.predict(X_test)

    accuracy = classifier.score(X_test,Y_test)
    a.append(accuracy)
    
plt.plot(range(2,40),a)
plt.show()
```

```python
a = []
for i in range(1,40):
    classifier = DecisionTreeClassifier(criterion='entropy',max_depth=i, random_state=0)
    classifier.fit(X_train,Y_train)
    pred_y = classifier.predict(X_test)

    accuracy = classifier.score(X_test,Y_test)
    a.append(accuracy)
    
plt.plot(range(1,40),a)
plt.show()
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', min_samples_split=29, max_depth=4,random_state=0)
classifier.fit(X_train,Y_train)
pred_y = classifier.predict(X_test)
accuracy = classifier.score(X_test,Y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,pred_y)

print(cm)
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set = X_test, Y_test
X1,X2 = np.meshgrid(np.arange(x_set[:,0].min()-1,x_set[:,0].max()+1,step=0.01),
                    np.arange(x_set[:,1].min()-1,x_set[:,1].max()+1,step=0.01))

plt.contourf(X1,X2,classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha= 0.75,
             cmap= ListedColormap(('red','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1], c= ListedColormap(('red','green'))(i),label=j)

plt.title('Decision Tree')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
